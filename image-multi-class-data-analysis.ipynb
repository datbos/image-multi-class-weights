### Variable Correllation Matrix
#### The cross correlation matrix indicates that there is a strong correlation with animals that have been vaccinated and animals that have been dewormed. There is a weaker correlation with animals that have bee sterilized.

```python
column_list1 = train.columns.values.tolist()
cmap = sns.diverging_palette(220,10, as_cmap=True)
correlations1 = train[column_list1].corr()
print ("Correlation matrix")
print (sns.heatmap(correlations1, cmap=cmap))
```
